```python
# Dependencies
import pandas as pd
import numpy as np
import requests
import json

# Other imports
from api_key import geoapify_key
```

```python
den_zipcodes = pd.read_csv("Denver_zips.csv")
den_zipcodes.head()
```

```python
# Set the geographical coordinates
latitude = 39.742043
longitude = -104.991531

# Set the parameters for the type of search
radius = 6500
limit = 20
filters = f"circle:{longitude},{latitude},{radius}"
bias = f"proximity:{longitude},{latitude}"

# set up a parameters dictionary
params = {
    "categories": "healthcare",
    "limit": limit,
    "filter":filters,
    "bias":bias,
    "apiKey":geoapify_key
}

# Set base URL
base_url = "https://api.geoapify.com/v2/places"
```

```python
# Empty lists for postcodes, facility name, lat, lng:
zip_codes = []
# facility_name = []
fac_type = []
# lat = []
# lng = []

print("Retrieving data...")
response = requests.get(base_url, params=params).json()
# print(json.dumps(response, indent=4, sort_keys=True))
# print(response["features"][0]["properties"]["name"])

facility = []

for feature in response["features"]:
    facility.append(feature["properties"]["address_line1"])
    zip_codes.append(feature["properties"]["postcode"])
    fac_type.append(feature["properties"]["categories"])
    
# This prints out (when limit is 20) 20 results
# print(json.dumps(response['features'], indent=4, sort_keys=True))
    
print(zip_codes)
print(facility)
print(fac_type)
```

```python
results_df = pd.DataFrame({"Zip Code": zip_codes,
               "Facility Name": facility,
               "Type": fac_type
})
results_df
```
